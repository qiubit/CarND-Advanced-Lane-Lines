```python
import src.calibrate as calibrate
import src.threshold as threshold
import src.transform as transform

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Camera calibration

```python
CALIB_IMG = 'camera_cal/calibration1.jpg'
```

```python
K, D = calibrate.calibrate_camera('camera_cal', nx=9, ny=6)
```

```python
img_cv = cv2.imread(CALIB_IMG)
undist_cv = calibrate.undistort(img_cv, K, D)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(img_cv)
ax1.set_title('Original', fontsize=50)
ax2.imshow(undist_cv)
ax2.set_title('Undistorted', fontsize=50)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.)
plt.savefig('output_images/camera_cal.jpg')
```
